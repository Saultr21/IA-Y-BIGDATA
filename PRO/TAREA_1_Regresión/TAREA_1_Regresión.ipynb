```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Datos del archivo adjunto
x = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26]).reshape((-1, 1))
y = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])

# Asigna todos los datos a los conjuntos de entrenamiento
x_train = x
y_train = y

# Crea el objeto de regresin lineal
regr = linear_model.LinearRegression()

# Entrena el modelo utilizando los conjuntos de entrenamiento
regr.fit(x_train, y_train)

# Realiza predicciones utilizando todo el conjunto de datos
y_pred = regr.predict(x)  # Predicciones para todo el conjunto de datos




# Coeficientes
print("Coeficientes: \n", regr.coef_)
# Calcula el error cuadrtico medio
print("Error cuadrtico medio: %.2f" % mean_squared_error(y, y_pred))
# Calcula el coeficiente de determinacin: 1 es una prediccin perfecta
print("Coeficiente de determinacin: %.2f" % r2_score(y, y_pred))


# Grafica los resultados
plt.scatter(x, y, color="black", label='Datos')
plt.plot(x, y_pred, color="blue", linewidth=3, label='Recta de regresin')  # Lnea de regresin en todo el rango
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regresin Lineal")
plt.legend()
plt.show()
```

## Frmulas:

### Pendiente \(b_1\):

$b_1 = \frac{\sum x_i y_i - \frac{\sum x_i \sum y_i}{n}}{\sum x_i^2 - \frac{(\sum x_i)^2}{n}}$

Esta frmula tambin puede expresarse como:

$b_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2} = \frac{S_{XY}}{S_{x}^2}$

Donde:

- $S_{XY}$ es la covarianza entre $x$ e $y$,
- $S_{x}^2$ es la varianza de $x$.

### Interseccin \(b_0\):

$b_0 = \bar{y} - b_1 \bar{x}$

Donde:

- $\bar{x}$ es la media de $x$,
- $\bar{y}$ es la media de $y$.

```python
import numpy as np  # Importa la biblioteca NumPy para trabajar con arreglos y realizar operaciones matemticas
import matplotlib.pyplot as plt  # Importa Matplotlib para crear grficos

# Datos de entrada
x = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26])  # Arreglo de valores independientes (xi)
y = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])  # Arreglo de valores dependientes (yi)

# Nmero de puntos
n = len(x)  # Calcula la cantidad de puntos (n) en los datos

# Sumas necesarias para las frmulas de regresin
sum_x = np.sum(x)  # Suma de todos los valores de x
sum_y = np.sum(y)  # Suma de todos los valores de y
sum_xy = np.sum(x * y)  # Suma del producto de cada par de valores x e y
sum_x_squared = np.sum(x ** 2)  # Suma de los cuadrados de cada valor de x

# Clculo de las medias de x e y
mean_x = np.mean(x)  # Media (promedio) de los valores de x
mean_y = np.mean(y)  # Media (promedio) de los valores de y

# Clculo de b1 (pendiente de la recta de regresin)
# La frmula para b1 es:
# b1 = ((xy) - (x * y) / n) / ((x^2) - (x)^2 / n)
b1 = (sum_xy - (sum_x * sum_y) / n) / (sum_x_squared - (sum_x ** 2) / n)

# Clculo de b0 (interseccin con el eje y)
# La frmula para b0 es:
# b0 = media_y - b1 * media_x
b0 = mean_y - b1 * mean_x

# Mostrar los coeficientes de la regresin lineal
print(f"Pendiente (b1): {b1}")  # Imprime el valor de la pendiente
print(f"Interseccin (b0): {b0}")  # Imprime el valor de la interseccin

# Clculo de los valores predichos de y usando la ecuacin de la recta de regresin
y_pred = b0 + b1 * x  # Predicciones de y basadas en la recta de regresin

# Graficar los puntos de datos y la recta de regresin
plt.scatter(x, y, color="black", label="Datos")  # Grafica los puntos de datos en negro
plt.plot(x, y_pred, color="blue", label="Recta de regresin")  # Grafica la recta de regresin en azul
plt.xlabel("xi")  # Etiqueta para el eje x
plt.ylabel("yi")  # Etiqueta para el eje y
plt.legend()  # Muestra la leyenda en el grfico
plt.show()  # Muestra el grfico
```

### Explicacin:
1. **Clculo de la pendiente (\(b_1\))**:
   - Usamos la frmula para \(b_1\), que relaciona la covarianza entre \(x\) e \(y\) con la varianza de \(x\). Este valor determina la inclinacin de la recta.
   
2. **Clculo de la interseccin (\(b_0\))**:
   - Despus de obtener \(b_1\), calculamos \(b_0\) usando la media de \(x\) e \(y\). Este valor representa dnde la recta corta el eje \(y\).

3. **Graficacin**:
   - Dibujamos los puntos proporcionados (\(x_i, y_i\)) como un diagrama de dispersin.
   - Luego graficamos la lnea de regresin ajustada, que representa la mejor estimacin lineal de \(y\) en funcin de \(x\).

```python
import numpy as np

# Datos de entrada
x = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26])
y = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])

# Clculos previos del ejercicio 2
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x_squared = np.sum(x ** 2)

mean_x = np.mean(x)
mean_y = np.mean(y)

# Coeficientes de la recta de regresin
b1 = (sum_xy - (sum_x * sum_y) / n) / (sum_x_squared - (sum_x ** 2) / n)
b0 = mean_y - b1 * mean_x

# Valores predichos de y
y_pred = b0 + b1 * x

# Clculo de R^2
ss_total = np.sum((y - mean_y) ** 2)  # Suma de cuadrados total
ss_residual = np.sum((y - y_pred) ** 2)  # Suma de cuadrados de los residuos
r_squared = 1 - (ss_residual / ss_total)

# Clculo del coeficiente de correlacin r
r = np.sqrt(r_squared)  # Si la relacin es positiva

# Mostrar resultados
print(f"Coeficiente de determinacin (R^2): {r_squared:.4f}")
print(f"Coeficiente de correlacin (r): {r:.4f}")
```

**Explicacin de cada parte:**

- Clculo de la recta de regresin: Como ya se hizo en el ejercicio 2, calculamos b0 y b1 y luego obtenemos y_pred.

- Clculo de $r^2$
 : Calculamos la suma de cuadrados total (ss_total) y la suma de cuadrados residuales (ss_residual) para obtener el coeficiente
$r^2$.

- Coeficiente de correlacin
 : Obtenemos

r tomando la raz cuadrada de
$r^2$
 , ya que estamos trabajando con una relacin lineal positiva entre x e y.
