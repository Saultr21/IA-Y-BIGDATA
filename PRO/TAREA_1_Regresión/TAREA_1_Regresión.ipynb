```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Datos del archivo adjunto
x = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26]).reshape((-1, 1))
y = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])

# Asigna todos los datos a los conjuntos de entrenamiento
x_train = x
y_train = y

# Crea el objeto de regresión lineal
regr = linear_model.LinearRegression()

# Entrena el modelo utilizando los conjuntos de entrenamiento
regr.fit(x_train, y_train)

# Realiza predicciones utilizando todo el conjunto de datos
y_pred = regr.predict(x)  # Predicciones para todo el conjunto de datos




# Coeficientes
print("Coeficientes: \n", regr.coef_)
# Calcula el error cuadrático medio
print("Error cuadrático medio: %.2f" % mean_squared_error(y, y_pred))
# Calcula el coeficiente de determinación: 1 es una predicción perfecta
print("Coeficiente de determinación: %.2f" % r2_score(y, y_pred))


# Grafica los resultados
plt.scatter(x, y, color="black", label='Datos')
plt.plot(x, y_pred, color="blue", linewidth=3, label='Recta de regresión')  # Línea de regresión en todo el rango
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regresión Lineal")
plt.legend()
plt.show()
```

## Fórmulas:

### Pendiente \(b_1\):

$b_1 = \frac{\sum x_i y_i - \frac{\sum x_i \sum y_i}{n}}{\sum x_i^2 - \frac{(\sum x_i)^2}{n}}$

Esta fórmula también puede expresarse como:

$b_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2} = \frac{S_{XY}}{S_{x}^2}$

Donde:

- $S_{XY}$ es la covarianza entre $x$ e $y$,
- $S_{x}^2$ es la varianza de $x$.

### Intersección \(b_0\):

$b_0 = \bar{y} - b_1 \bar{x}$

Donde:

- $\bar{x}$ es la media de $x$,
- $\bar{y}$ es la media de $y$.

```python
import numpy as np  # Importa la biblioteca NumPy para trabajar con arreglos y realizar operaciones matemáticas
import matplotlib.pyplot as plt  # Importa Matplotlib para crear gráficos

# Datos de entrada
x = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26])  # Arreglo de valores independientes (xi)
y = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])  # Arreglo de valores dependientes (yi)

# Número de puntos
n = len(x)  # Calcula la cantidad de puntos (n) en los datos

# Sumas necesarias para las fórmulas de regresión
sum_x = np.sum(x)  # Suma de todos los valores de x
sum_y = np.sum(y)  # Suma de todos los valores de y
sum_xy = np.sum(x * y)  # Suma del producto de cada par de valores x e y
sum_x_squared = np.sum(x ** 2)  # Suma de los cuadrados de cada valor de x

# Cálculo de las medias de x e y
mean_x = np.mean(x)  # Media (promedio) de los valores de x
mean_y = np.mean(y)  # Media (promedio) de los valores de y

# Cálculo de b1 (pendiente de la recta de regresión)
# La fórmula para b1 es:
# b1 = (Σ(xy) - (Σx * Σy) / n) / (Σ(x^2) - (Σx)^2 / n)
b1 = (sum_xy - (sum_x * sum_y) / n) / (sum_x_squared - (sum_x ** 2) / n)

# Cálculo de b0 (intersección con el eje y)
# La fórmula para b0 es:
# b0 = media_y - b1 * media_x
b0 = mean_y - b1 * mean_x

# Mostrar los coeficientes de la regresión lineal
print(f"Pendiente (b1): {b1}")  # Imprime el valor de la pendiente
print(f"Intersección (b0): {b0}")  # Imprime el valor de la intersección

# Cálculo de los valores predichos de y usando la ecuación de la recta de regresión
y_pred = b0 + b1 * x  # Predicciones de y basadas en la recta de regresión

# Graficar los puntos de datos y la recta de regresión
plt.scatter(x, y, color="black", label="Datos")  # Grafica los puntos de datos en negro
plt.plot(x, y_pred, color="blue", label="Recta de regresión")  # Grafica la recta de regresión en azul
plt.xlabel("xi")  # Etiqueta para el eje x
plt.ylabel("yi")  # Etiqueta para el eje y
plt.legend()  # Muestra la leyenda en el gráfico
plt.show()  # Muestra el gráfico
```

### Explicación:
1. **Cálculo de la pendiente (\(b_1\))**:
   - Usamos la fórmula para \(b_1\), que relaciona la covarianza entre \(x\) e \(y\) con la varianza de \(x\). Este valor determina la inclinación de la recta.
   
2. **Cálculo de la intersección (\(b_0\))**:
   - Después de obtener \(b_1\), calculamos \(b_0\) usando la media de \(x\) e \(y\). Este valor representa dónde la recta corta el eje \(y\).

3. **Graficación**:
   - Dibujamos los puntos proporcionados (\(x_i, y_i\)) como un diagrama de dispersión.
   - Luego graficamos la línea de regresión ajustada, que representa la mejor estimación lineal de \(y\) en función de \(x\).

```python
import numpy as np

# Datos de entrada
x = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26])
y = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])

# Cálculos previos del ejercicio 2
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x_squared = np.sum(x ** 2)

mean_x = np.mean(x)
mean_y = np.mean(y)

# Coeficientes de la recta de regresión
b1 = (sum_xy - (sum_x * sum_y) / n) / (sum_x_squared - (sum_x ** 2) / n)
b0 = mean_y - b1 * mean_x

# Valores predichos de y
y_pred = b0 + b1 * x

# Cálculo de R^2
ss_total = np.sum((y - mean_y) ** 2)  # Suma de cuadrados total
ss_residual = np.sum((y - y_pred) ** 2)  # Suma de cuadrados de los residuos
r_squared = 1 - (ss_residual / ss_total)

# Cálculo del coeficiente de correlación r
r = np.sqrt(r_squared)  # Si la relación es positiva

# Mostrar resultados
print(f"Coeficiente de determinación (R^2): {r_squared:.4f}")
print(f"Coeficiente de correlación (r): {r:.4f}")
```

**Explicación de cada parte:**

- Cálculo de la recta de regresión: Como ya se hizo en el ejercicio 2, calculamos b0 y b1 y luego obtenemos y_pred.

- Cálculo de $r^2$
 : Calculamos la suma de cuadrados total (ss_total) y la suma de cuadrados residuales (ss_residual) para obtener el coeficiente
$r^2$.
